```python
def stats_test(col):
  if stroke_dataset[col].dtype == float:
    tmp0 = stroke_dataset.loc[stroke_dataset['stroke'] == 0, col].values
    tmp1 = stroke_dataset.loc[stroke_dataset['stroke'] == 1, col].values

    result = stats.ttest_ind(tmp0, tmp1)[0:2]

  else:
    contingency_table = stroke_dataset.pivot_table(index = [col], columns = 'stroke', values = 'age', aggfunc = 'count')
    contingency_table = contingency_table.fillna(0)
    contingency_table = contingency_table.values

    result = stats.chi2_contingency(contingency_table, correction = False)[0:2]

  return result
```

```python
train_data = lgb.Dataset(encoded_X_train, label=y_train)
test_data = lgb.Dataset(encoded_X_test, label=y_test, reference=train_data)

params = {
    'objective': 'binary',
    'metric': 'auc',
    'is_unbalance': 'true',
    'boosting_type': 'gbdt',
    'num_leaves': 40,  # 增加葉子數
    'max_depth': -1,
    'learning_rate': 0.01,  # 減小學習率
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0,
    'lambda_l1': 0.1,  # L1 正則化
    'lambda_l2': 0.1,  # L2 正則化
    'min_data_in_leaf': 30,  # 增加每個葉子最少的數據樣本數
    'max_bin': 255  # 增加分割數據的箱數
}

num_round = 100
lgb_model = lgb.train(params, train_data, num_round, valid_sets=[train_data, test_data])

y_pred = lgb_model.predict(encoded_X_test, num_iteration=lgb_model.best_iteration)
y_pred_binary = [1 if pred > 0.5 else 0 for pred in y_pred]

# 評估模型性能
accuracy = accuracy_score(y_test, y_pred_binary)
conf_matrix = confusion_matrix(y_test, y_pred_binary)
class_report = classification_report(y_test, y_pred_binary)
auc_score = roc_auc_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
print("AUC (LightGBM):", auc_score)
```
